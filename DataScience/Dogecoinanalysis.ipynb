```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
doge_df=pd.read_csv('DOGE-USD.csv')
doge_df
```

```python
doge_df.columns
```

```python
doge_df.drop(['Adj Close','Volume'],axis=1)
```

```python
doge_df_top=doge_df.head(15)
doge_df_top
```

```python
doge_df_top.drop(['Adj Close','Volume'],axis=1)
```

```python
plt.figure(figsize=(20,10))
plt.plot(doge_df_top.Date,doge_df_top.High,marker='o',ls='-',lw=3,ms=12,mew=1,alpha=.7)
plt.plot(doge_df_top.Date,doge_df_top.Low,marker='o',ls='-',lw=3,ms=12,mew=1,alpha=.9)
plt.plot(doge_df_top.Date,doge_df_top.Open,marker='o',ls='-',lw=3,ms=12,mew=1,alpha=.7)
plt.plot(doge_df_top.Date,doge_df_top.Close,marker='d',ls='-',lw=3,ms=12,mew=1,alpha=.7)
# plt.plot(final_df.Date,final_df.Volume,marker='d',ls='-',lw=3,ms=12,mew=1,alpha=.7)
plt.xlabel('Date',fontsize = 25)
plt.ylabel('USD',fontsize = 25)
plt.grid(color='g', linestyle='-', linewidth=0.8,alpha=0.5)
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.title("Dogecoin Market In First 15 Days",fontsize = 30,pad=20);
plt.legend(['High','Low','Open','Close']);
```
