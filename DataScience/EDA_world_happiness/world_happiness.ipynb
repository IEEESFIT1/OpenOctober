# World Happiness Data

This dataset contains Happiness Report of 156 countries. The parameters taken into account are:
- Score : The happiness score the country has.
- GDP per capita : Per capita gross domestic product (GDP) is a metric that breaks down a country's economic output per person and is calculated by dividing the GDP of a country by its population.
- Social Support : Social support is the perception and actuality that one is cared for, has assistance available from other people, and most popularly, that one is part of a supportive social network. These supportive resources can be emotional, informational, or companionship; tangible or intangible.
- Healthy life expectancy : The Healthy Life Years indicator, also known as disability-free life expectancy or Sullivan's Index, is a European structural indicator computed by Eurostat. It is one of the summary measures of population health, known as health expectancies, composite measures of health that combine mortality and morbidity data to represent overall population health on a single indicator. HLY measures the number of remaining years that a person of a certain age is expected to live without disability.
- Freedom to make life choices : Freedom of choice describes an individual's opportunity and autonomy to perform an action selected from at least two available options, unconstrained by external parties.
- Generosity : Generosity is the virtue of being liberal in giving, often as gifts. Generosity is regarded as a virtue by various world religions, and is often celebrated in cultural and religious ceremonies.
- Perceptions of corruption : The Corruption Perceptions Index is an index published annually by Transparency International since 1995 which ranks countries "by their perceived levels of public sector corruption, as determined by expert assessments and opinion surveys." The CPI generally defines corruption as "the misuse of public power for private benefit".

The dataset was taken from Kaggle. You can find it [here](https://www.kaggle.com/unsdsn/world-happiness).

```python
#EXAMPLE OF A LINE GRAPH TO TRY TO FIND A RELATION BETWEEN THE TWO LABELS
plt.plot(data_df['Overall rank'],data_df['GDP per capita']);
plt.xlabel('Overall rank')
plt.ylabel('GDP per capita')
plt.title('GDP per capita vs Overall rank')
```

There is no clear relation between GDP per capita and the overall rank of happiness for a country. There are cases wehre the GDP per capita is high but the overall rank is low- like observing the point between 140 and 160 where the GDP per capita is comparable to the countries with a higher rank( low numeric number ).


Since the line plot comes out to be highly variable without any pattern, let us look at the scatter plot to find the better representation of the data.

```python
#TRYING TO FIND A PATTERN.
plt.figure(figsize=(12, 6))
sns.scatterplot(data_df['Overall rank'], data_df['GDP per capita'],hue=data_df['GDP per capita']);
```


A scatter plot is similar to a line graph, but with just the points. We can use a hue colour to better visualise the data depending on a paramter. The points with a darker shade of purple have a higher GDP per capita.

Now, let us use a for loop for finding the line graph and scatter plots for all the seven labels that we have.


```python
fig, axes = plt.subplots(4, 2, figsize=(20, 20))

k = 2
for i in range(0,4):
  for j in range(0,2):
    if i is 3 and j is 1:
      break;
    axes[i,j].plot(data_df['Overall rank'],data_df[data_df.columns[k]],color="purple")
    axes[i,j].set_xlabel('Overall rank')
    axes[i,j].set_ylabel(data_df.columns[k])
    k+=1
   
    plt.tight_layout(pad=2);
```

## Asking and Answering Questions



Finding the top ten countries in each label.

```python
fig, axes = plt.subplots(nrows=3, ncols=2,constrained_layout=True,figsize=(12,8))

sns.barplot(x='GDP per capita',y='Country or region',data=data_df.nlargest(10,'GDP per capita'),ax=axes[0,0],palette="Spectral")

sns.barplot(x='Social support' ,y='Country or region',data=data_df.nlargest(10,'Social support'),ax=axes[0,1],palette="coolwarm")

sns.barplot(x='Healthy life expectancy' ,y='Country or region',data=data_df.nlargest(10,'Healthy life expectancy'),ax=axes[1,0],palette='cubehelix')

sns.barplot(x='Freedom to make life choices' ,y='Country or region',data=data_df.nlargest(10,'Freedom to make life choices'),ax=axes[1,1],palette='rocket')


sns.barplot(x='Generosity' ,y='Country or region',data=data_df.nlargest(10,'Generosity'),ax=axes[2,0],palette='magma')

sns.barplot(x='Perceptions of corruption' ,y='Country or region',data=data_df.nlargest(10,'Perceptions of corruption'),ax=axes[2,1],palette='viridis')
```

**Q1-DOES THE GDP CHANGE THE HEALTHY LIFE EXPECTANCY?**

It is evident from the graph that GDP does affect the healthy life expectancy. The higher the GDP the higher the life expectancy.

```python
data = data_df.nlargest(10,'Healthy life expectancy')
plt.plot(data['Healthy life expectancy'],data['GDP per capita']);
plt.xlabel('Healthy life expectancy')
plt.ylabel('GDP per capita')
plt.title('GDP per capita vs Healthy Life Expectancy')
```

**Q2-HOW MANY COUNTRIES ARE ABOVE THE AVERAGE SCORE?**

```python
data_df.describe()
```

This shows the mean or average happiness score is 5.4076096.
Let us find the number of countries above or below this score.

```python
above_mean = data_df[data_df['Score']>5.407096]
below_mean = data_df[data_df['Score']<5.407096] 
```

```python
import matplotlib.pyplot as plt

labels = ['Above mean','Below mean']
sizes = [above_mean.shape[0],below_mean.shape[0]]
colors = ['yellowgreen', 'lightcoral']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**Q3-HOW MANY COUNTRIES LIE IN THE UPPER QUARTILE OF THE HAPPINESS SCORE?**

Looking at the top percentage of countries who are happy.

```python
#from the describe table the upper quartile range begins at 6.18450
above_upperQ = data_df[data_df['Score']>6.18450]
below_upperQ = data_df[data_df['Score']<6.18450] 
```

```python
labels = ['Above Upper Quartile','Below Upper Quartile']
sizes = [above_upperQ.shape[0],below_upperQ.shape[0]]
colors = ['seagreen', 'sandybrown']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**Q4-FINDING CORRELATION BETWEEN THE OVERALL RANK FOR HAPPINESS AND THE FACTORS CONTRIBUTING TO IT.**

```python
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(data_df.corr(),annot=True,linewidths=.5,cmap='cubehelix',ax=ax)
```

This heatmap shows a positive correlation between the factors- score, GDP per capita, Social Support and Healthy Life expectancy.

**Q5-DOES THE GEOGRAPHICAL DEMOGRAPHIC AFFECT THE OVERALL HAPPINESS SCORE?**

It seems that the overall happiness score is related to the geographical demographics in the following way:

- The western world countries rank the highest.
- The countries in Oceania- Austrailia and New Zealand follow the happiness score after the western world.
- Europe and The Kingdom of Saudi Arabia follows the Oceania.
- Asia ranks lower in the happiness score. The developing countries ranked lower. 


```python
import plotly.graph_objs as go
from plotly.offline import iplot

data = dict(type = 'choropleth', 
           locations = data_df['Country or region'],
           locationmode = 'country names',
           colorscale='magma',
           z = data_df['Score'], 
           text = data_df['Country or region'],
           colorbar = {'title':'Happiness Score'})

layout = dict(title = 'Visualisation Map (Happiness Score)', 
              geo = dict(showframe = True, projection = {'type': 'azimuthal equal area'}))

choromap3 = go.Figure(data = [data], layout=layout)
iplot(choromap3)
```
